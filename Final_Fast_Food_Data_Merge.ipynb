```python
from google.colab import files
import pandas as pd
from google.colab import drive
import os
```

```python
df_ff = pd.read_csv("FastFoodRestaurants.csv")

# Drop duplicates
df_ff.drop_duplicates(inplace=True)

# Drop unnecessary columns
df_ff = df_ff.drop(columns=['country', 'keys', 'latitude', 'longitude', 'websites'])

# Drop rows with null values
df_cleanedff = df_ff.dropna()

# New column: count of fast food restaurants by postal code
fast_food_counts = df_cleanedff.groupby('postalCode').size().reset_index(name='fast_food_count')
df_ffc = pd.DataFrame(fast_food_counts)

# Rename the column to 'Zip Code' if necessary
df_ffc.rename(columns={'postalCode': 'Zip Code'}, inplace=True)

# Save the cleaned fast food data to Google Drive
drive.mount('/content/drive')  # Mount Google Drive
os.makedirs('/content/drive/My Drive/Final_ML', exist_ok=True)
df_ffc.to_csv('/content/drive/My Drive/Final_ML/Cleaned_Fast_Food_Restaurants.csv', index=False)

# Step 2: Clean Diabetes Data
# Upload the Diabetes data
df_da = pd.read_csv("DiabetesAtlasData.csv")

# Drop duplicates
df_da.drop_duplicates(inplace=True)

# Dictionary to map full state names to their abbreviations
state_abbreviations = {
    'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR', 'California': 'CA',
    'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE', 'Florida': 'FL', 'Georgia': 'GA',
    'Hawaii': 'HI', 'Idaho': 'ID', 'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA',
    'Kansas': 'KS', 'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD',
    'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS', 'Missouri': 'MO',
    'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV', 'New Hampshire': 'NH', 'New Jersey': 'NJ',
    'New Mexico': 'NM', 'New York': 'NY', 'North Carolina': 'NC', 'North Dakota': 'ND', 'Ohio': 'OH',
    'Oklahoma': 'OK', 'Oregon': 'OR', 'Pennsylvania': 'PA', 'Rhode Island': 'RI', 'South Carolina': 'SC',
    'South Dakota': 'SD', 'Tennessee': 'TN', 'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT',
    'Virginia': 'VA', 'Washington': 'WA', 'West Virginia': 'WV', 'Wisconsin': 'WI', 'Wyoming': 'WY'
}

# Replace full state names with abbreviations
df_da['State'] = df_da['State'].replace(state_abbreviations)

df_da['County'] = df_da['County'] + ', ' + df_da['State']

# Rename the column to 'Zip Code' if necessary
df_da.rename(columns={'County_FIPS': 'Zip Code'}, inplace=True)

# Save the cleaned diabetes data to Google Drive
df_da.to_csv('/content/drive/My Drive/Final_ML/Cleaned_Diabetes_Atlas.csv', index=False)

# Step 3: Clean Food Access Data
# Upload the Food Access data
df = pd.read_csv("food_access.csv")

# Drop duplicates and rows with null values
df.drop_duplicates(inplace=True)
df_fa = df.dropna()

# Drop irrelevant columns
df_fa = df_fa.drop(columns=[
    'Housing Data.Residing in Group Quarters',
    'Housing Data.Total Housing Units',
    'Vehicle Access.1 Mile',
    'Vehicle Access.1/2 Mile',
    'Vehicle Access.10 Miles',
    'Vehicle Access.20 Miles',
    'Low Access Numbers.Children.1 Mile',
    'Low Access Numbers.Children.1/2 Mile',
    'Low Access Numbers.Children.10 Miles',
    'Low Access Numbers.Children.20 Miles',
    'Low Access Numbers.Low Income People.1 Mile',
    'Low Access Numbers.Low Income People.1/2 Mile',
    'Low Access Numbers.Low Income People.10 Miles',
    'Low Access Numbers.Low Income People.20 Miles',
    'Low Access Numbers.People.1 Mile',
    'Low Access Numbers.People.1/2 Mile',
    'Low Access Numbers.People.10 Miles',
    'Low Access Numbers.People.20 Miles',
    'Low Access Numbers.Seniors.1 Mile',
    'Low Access Numbers.Seniors.1/2 Mile',
    'Low Access Numbers.Seniors.10 Miles',
    'Low Access Numbers.Seniors.20 Miles'
])

# Dictionary to map full state names to their abbreviations
state_abbreviations = {
    'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR', 'California': 'CA',
    'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE', 'Florida': 'FL', 'Georgia': 'GA',
    'Hawaii': 'HI', 'Idaho': 'ID', 'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA',
    'Kansas': 'KS', 'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD',
    'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS', 'Missouri': 'MO',
    'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV', 'New Hampshire': 'NH', 'New Jersey': 'NJ',
    'New Mexico': 'NM', 'New York': 'NY', 'North Carolina': 'NC', 'North Dakota': 'ND', 'Ohio': 'OH',
    'Oklahoma': 'OK', 'Oregon': 'OR', 'Pennsylvania': 'PA', 'Rhode Island': 'RI', 'South Carolina': 'SC',
    'South Dakota': 'SD', 'Tennessee': 'TN', 'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT',
    'Virginia': 'VA', 'Washington': 'WA', 'West Virginia': 'WV', 'Wisconsin': 'WI', 'Wyoming': 'WY'
}

# Replace full state names with abbreviations
df_fa['State'] = df_fa['State'].replace(state_abbreviations)

df_fa['County'] = df_fa['County'] + ', ' + df_fa['State']

df_fa = df_fa.drop(columns=['State'])

# Save the cleaned Food Access data to Google Drive
df_fa.to_csv('/content/drive/My Drive/Final_ML/Cleaned_Food_Access.csv', index=False)

```

```python
# Step 4: Merge DataFrames
# Merge the Food Access data with the Diabetes data by 'County'
merged_df = df_fa.merge(df_da, on='County', how='inner')
print(f"After merging Food Access and Diabetes: {merged_df.shape}")

# Merge the result with the Fast Food data by 'Zip Code'
merged_df = merged_df.merge(df_ffc[['Zip Code', 'fast_food_count']], on='Zip Code', how='inner')
print(f"After merging with Fast Food data: {merged_df.shape}")

# Handle duplicate 'Zip Code' columns (if present)
if 'Zip Code_x' in merged_df.columns and 'Zip Code_y' in merged_df.columns:
    merged_df['Zip Code'] = merged_df['Zip Code_x'].fillna(merged_df['Zip Code_y'])
    merged_df.drop(columns=['Zip Code_x', 'Zip Code_y'], inplace=True)

# Drop rows with missing values
merged_df.dropna(inplace=True)

# Calculate the concentration of fast food restaurants
merged_df['fast_food_concentration'] = merged_df['fast_food_count'] / merged_df['Population']

# Step 7: Save the final merged dataset to Google Drive
merged_df.to_csv('/content/drive/My Drive/Final_ML/Merged_Food_Health_Data.csv', index=False)

# Inspect the final dataset
print(merged_df.head())
```
